# Subset Selection for All Observations and Combined with Log Reg and XGboost. 

This document presents the results of performing subset selection to determine the best percent development and train/test split for Logistic Regression and XGBoost models. 

## Data

Two sources were used for each model:

*site_species_presence.csv* -> All observations regardless of repeated sites across years with only observations with empty watershed area data removed. (570 sites)

*site_species_presence_combined.csv* -> Observations of the same site across years are combined and observations with empty watershed area data are removed. (468 sites)



## Subset Selection Process and Results

The following combinations were done using first site-grouped cross validation and then stratified cross validation to determine the best train/test split for each threshold of percent development: 

- Not-combined observations, with all fish columns
- Combined observations, with all fish columns
- Not-combined observations, with no fish columns
- Combined observations, with no fish columns
